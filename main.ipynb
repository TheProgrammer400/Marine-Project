## Architecture

```mathematics

Input Image
   ↓
Backbone
   ↓
Neck
   ↓
Detection Head
   ↓
Bounding Boxes + Confidence
   ↓
Detected shrimp boxes
   ↓
Crop each shrimp
   ↓
Backbone + Regression Head
   ↓
Mass per shrimp
```


`Backbone` is output of contrastive learning
